```python
import joblib
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


# Gender sÃ¼tunu iÃ§in Gender_M oluÅŸtur
data['Gender_M'] = data['Gender'].apply(lambda x: 1 if x == 'M' else 0)

# EÄŸitim Seviyesi (Education_Level) iÃ§in Ordinal Encoding
education_order = ['Unknown', 'Uneducated', 'High School', 'Graduate', 'College', 'Post-Graduate', 'Doctorate']
education_mapping = {label: idx for idx, label in enumerate(education_order)}
data['Education_Level'] = data['Education_Level'].map(education_mapping)
joblib.dump(education_mapping, 'education_mapping.pkl')

# Gelir Kategorisi (Income_Category) iÃ§in Ordinal Encoding
income_order = ['Unknown', 'Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +']
income_mapping = {label: idx for idx, label in enumerate(income_order)}
data['Income_Category'] = data['Income_Category'].map(income_mapping)
joblib.dump(income_mapping, 'income_mapping.pkl')

# Kart Kategorisi (Card_Category) iÃ§in Ordinal Encoding
card_order = ['Blue', 'Silver', 'Gold', 'Platinum']
card_mapping = {label: idx for idx, label in enumerate(card_order)}
data['Card_Category'] = data['Card_Category'].map(card_mapping)
joblib.dump(card_mapping, 'card_mapping.pkl')

# Meslek (jobs) iÃ§in Label Encoding
le_jobs = LabelEncoder()
data['jobs'] = le_jobs.fit_transform(data['jobs'])

job_mapping = dict(zip(le_jobs.classes_, le_jobs.transform(le_jobs.classes_)))
joblib.dump(job_mapping, 'job_mapping.pkl')

# Medeni Durum (Marital_Status) iÃ§in One-Hot Encoding
encoder_marital_status = OneHotEncoder(sparse_output =False, drop='first')
marital_status_encoded = encoder_marital_status.fit_transform(data[['Marital_Status']])
marital_status_encoded_df = pd.DataFrame(marital_status_encoded, columns=encoder_marital_status.get_feature_names_out(['Marital_Status']))
data = pd.concat([data, marital_status_encoded_df], axis=1)
data = data.drop(columns=['Marital_Status'])
joblib.dump(encoder_marital_status, 'encoder_marital_status.pkl')

# Object tipindeki sÃ¼tunlarÄ± sil
object_columns = data.select_dtypes(include=['object']).columns
data = data.drop(columns=object_columns)

# Ä°ÅŸlenmiÅŸ veriyi kontrol edin
print(data.info())
```

```python
def detect_outliers(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Numeric sÃ¼tunlarÄ± seÃ§
numeric_columns = data.select_dtypes(include=[np.number]).columns

num_columns = len(numeric_columns)
fig, axes = plt.subplots(nrows=num_columns, ncols=1, figsize=(10, 5 * num_columns))

for i, column in enumerate(numeric_columns):
    sns.boxplot(ax=axes[i], data=data, x=column, orient='h')
    lower_bound, upper_bound = detect_outliers(data[column])
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    num_outliers = len(outliers)
    axes[i].set_title(f'SÃ¼tun: {column}\n'
                      f'AykÄ±rÄ± deÄŸer sayÄ±sÄ±: {num_outliers}\n'
                      f'Alt sÄ±nÄ±r: {lower_bound:.2f}\n'
                      f'Ãœst sÄ±nÄ±r: {upper_bound:.2f}')

plt.tight_layout()
plt.show()
```

```python
def detect_outliers(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Belirli sÃ¼tunlar hariÃ§ tutulacak sÃ¼tunlarÄ±n isimlerini belirleyin
exclude_columns = ['Customer_Age', 'Card_Category','credit_card_tran_Amt','credit_card_tran_Amt_1',
                   'credit_card_tran_Amt_2','credit_card_tran_Amt_3','credit_card_tran_Amt_4',
                   'credit_card_tran_Amt_5','ocak_to_subat_tran_ct_degisim','credit_cart_tran_Amt_degisim_1to2']  

# Yeni DataFrame'in bir kopyasÄ±nÄ± oluÅŸturun
data_cleaned = data.copy()

# Numeric sÃ¼tunlarÄ± seÃ§
numeric_columns = data.select_dtypes(include=[np.number]).columns

# AykÄ±rÄ± deÄŸerleri temizleme iÅŸlemi
for column in numeric_columns:
    if column not in exclude_columns:
        lower_bound, upper_bound = detect_outliers(data[column])
        # AykÄ±rÄ± deÄŸerleri iÃ§ermeyen satÄ±rlarÄ± seÃ§
        data_cleaned = data_cleaned[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# SonuÃ§larÄ± gÃ¶rÃ¼ntÃ¼leyin
print(f"TemizlenmiÅŸ DataFrame'in boyutu: {data_cleaned.shape}")
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
import pandas as pd
import numpy as np

# Regresyon modelleri ve hiperparametreleri
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'RandomForestRegressor': RandomForestRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),
    'XGBRegressor': XGBRegressor(),
    'LGBMRegressor': LGBMRegressor(verbose=-1),
    'CatBoostRegressor': CatBoostRegressor(verbose=0),
    'KNeighborsRegressor': KNeighborsRegressor(),
    'MLPRegressor': MLPRegressor(),
    'AdaBoostRegressor': AdaBoostRegressor(),
    'ExtraTreesRegressor': ExtraTreesRegressor(),
    'BaggingRegressor': BaggingRegressor()
}

params = {
    'LinearRegression': {},
    'Ridge': {'alpha': [0.1, 1.0, 10.0]},
    'Lasso': {'alpha': [0.001, 0.01, 0.1, 1.0]},
    'ElasticNet': {'alpha': [0.1, 1.0], 'l1_ratio': [0.1, 0.5, 0.9]},
    'RandomForestRegressor': {
        'n_estimators': [100, 200, 300],
        'max_depth': [None, 10, 20, 30]
    },
    'GradientBoostingRegressor': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5]
    },
    'DecisionTreeRegressor': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'XGBRegressor': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5]
    },
    'LGBMRegressor': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 4, 5],
        'num_leaves': [31, 40, 50]
    },
    'CatBoostRegressor': {
        'iterations': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'depth': [3, 4, 5]
    },
    'KNeighborsRegressor': {
        'n_neighbors': [3, 5, 7, 9]
    },
    'MLPRegressor': {
        'hidden_layer_sizes': [(50,), (100,), (100,)],
        'activation': ['tanh', 'relu'],
        'solver': ['sgd', 'adam'],
        'alpha': [0.0001, 0.001, 0.01],
        'learning_rate': ['constant', 'adaptive']
    },
    'AdaBoostRegressor': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.05, 0.1]
    },
    'ExtraTreesRegressor': {
        'n_estimators': [100, 200, 300],
        'max_depth': [None, 10, 20, 30]
    },
    'BaggingRegressor': {
        'n_estimators': [10, 50, 100],
        'max_samples': [0.5, 0.7, 1.0],
        'max_features': [0.5, 0.7, 1.0]
    }
}

# Her hedef iÃ§in en iyi modeli ve performans metriklerini kaydetmek iÃ§in boÅŸ bir liste
all_results = []

for target in targets:
    y = targets[target]
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    results = []
    
    for model_name, model in models.items():
        print(f"Training {model_name} for {target}...")
        gs = GridSearchCV(model, params[model_name], scoring='r2', cv=5, verbose=1, n_jobs=-1)
        gs.fit(X_train, y_train)

        # En iyi modeli seÃ§me
        best_model = gs.best_estimator_

        # EÄŸitim seti Ã¼zerinde tahmin yapma
        y_train_pred = best_model.predict(X_train)
        y_test_pred = best_model.predict(X_test)

        # Metriklerin hesaplanmasÄ±
        train_r2 = r2_score(y_train, y_train_pred)
        test_r2 = r2_score(y_test, y_test_pred)
        train_mse = mean_squared_error(y_train, y_train_pred)
        test_mse = mean_squared_error(y_test, y_test_pred)
        train_mae = mean_absolute_error(y_train, y_train_pred)
        test_mae = mean_absolute_error(y_test, y_test_pred)

        results.append([target, model_name, train_r2, test_r2, train_mse, test_mse, train_mae, test_mae, gs.best_params_])
    
    all_results.extend(results)

# SonuÃ§larÄ± DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼rme
results_df = pd.DataFrame(all_results, columns=['Target', 'Model', 'Train_R2', 'Test_R2', 'Train_MSE', 'Test_MSE', 'Train_MAE', 'Test_MAE', 'Best_Params'])
print(results_df)
```

```python
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Hedeflerin hazÄ±rlanmasÄ±
target_names = {
    'Haziran_Trans_Amt': 'Haziran_Trans_Amt',
    'credit_card_tran_Amt_6': 'credit_card_tran_Amt_6',
    'Jun_hesap_bakiyesi': 'Jun_hesap_bakiyesi'
}

models = {
    'Haziran_Trans_Amt': ElasticNet(alpha=1.0, l1_ratio=0.9),
    'credit_card_tran_Amt_6': Lasso(alpha=1.0),
    'Jun_hesap_bakiyesi': ExtraTreesRegressor(max_depth=10, n_estimators=200)
}

# Modelleri eÄŸitme ve kaydetme
for target_key, model in models.items():
    target = targets[target_names[target_key]]
    
    # EÄŸitim ve test setlerine ayÄ±rma
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, target, test_size=0.2, random_state=42)
    
    # Modeli eÄŸitme
    model.fit(X_train, y_train)
    
    # Test seti Ã¼zerinde tahmin yapma
    y_test_pred = model.predict(X_test)
    
    # Test RÂ² skorunu hesaplama
    test_r2 = r2_score(y_test, y_test_pred)
    
    # Modeli kaydetme
    model_filename = f'{target_key}_model.pkl'
    joblib.dump(model, model_filename)
    
    # Test RÂ² skorunu yazdÄ±rma
    print(f'{target_key} iÃ§in model {model_filename} olarak kaydedildi. Test RÂ²: {test_r2:.4f}')
```
